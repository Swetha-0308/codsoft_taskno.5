**CREDIT CARD FRAUD DETECTION**

importing the dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#loading the dataset to a pandas DataFrame
data1=pd.read_csv('/content/creditcard.csv')
```

```python
#first 5 rows of the dataset
data1.head()
```

```python
data1.tail()
```

```python
#dataset information
data1.info()
```

```python
# checking the number of missing values
data1.isnull().sum()
```

```python
# distribution of legit transaction and fraudulent transactions
data1['Class'].value_counts()
```

this dataset is highly unblanced

0--> Normal Transaction
1-->fraudulent transaction

```python
#seperating the data for analysis
legit=data1[data1.Class==0]
fraud=data1[data1.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare the values for both transactions
data1.groupby('Class').mean()
```

Under-sampling

Build a sample dataset containing similar distribution of normal

Number of fraudulent transaction--->492

```python
legit_sample=legit.sample(n=492)
```

concatenating two DataFrame

```python
new_dataset=pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

splitting the data into features and target

```python
x=new_dataset.drop(columns='Class', axis=1)
y=new_dataset['Class']
```

```python
print(x)
```

```python
print(y)
```

spliting the data into training data and splitting data

```python
x_train, x_test, y_train, y_test= train_test_split(x,y, test_size=0.2, stratify=y, random_state=2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

Model training

logistic regression

```python
model= LogisticRegression()
```

```python
#training the Logistic Regression model with training data
model.fit(x_train, y_train)
```

Model evaluation

Accuracy score


```python
# accuracy on training data
x_train_prediction=model.predict(x_train)
training_data_accuracy=accuracy_score(x_train_prediction, y_train)
```

```python
print('Accuracy on Training data: ', training_data_accuracy)
```

```python
# accuracy on test data
x_test_prediction=model.predict(x_test)
test_data_accuracy=accuracy_score(x_test_prediction, y_test)
```

```python
print('Accuracy score on test data:' , test_data_accuracy)
```
